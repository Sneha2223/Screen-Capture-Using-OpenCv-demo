```python
import cv2
import pytesseract
import numpy as np
from mss import mss
import time

# Function to capture a single screenshot of the entire screen
def capture_full_screen():
    with mss() as sct:
        sct_img = sct.grab(sct.monitors[0])  # Capture the entire screen
        return cv2.cvtColor(np.array(sct_img), cv2.COLOR_RGBA2BGR)

# Function to save the screenshot
def save_screenshot(image, filename='screenshot.png'):
    cv2.imwrite(filename, image)

# Function to perform OCR on the screenshot
def perform_ocr(image_path):
    screenshot = cv2.imread(image_path)
    gray_image = cv2.cvtColor(screenshot, cv2.COLOR_BGR2GRAY)
    text = pytesseract.image_to_string(gray_image)
    return text

def main():
    while True:
        # Capture a single screenshot of the entire screen
        screenshot = capture_full_screen()

        # Save the screenshot
        save_screenshot(screenshot)

        # Perform OCR on the screenshot
        extracted_text = perform_ocr('screenshot.png')

        # Print the extracted text
        print("Extracted Text:")
        print(extracted_text)

        # Display the screenshot (optional)
        cv2.imshow('Live Screenshot', screenshot)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        time.sleep(0.1)  # A short delay to not overload the CPU

    cv2.destroyAllWindows()

if __name__ == '__main__':
    main()
```
